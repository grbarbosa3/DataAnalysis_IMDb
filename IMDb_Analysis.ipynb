# IMDb Analysis

### Libraries

```python
import pandas as pd
import datetime as dt
```

```python
pd.set_option('display.max_rows', None)
```

### Impoting data

```python
movies_df = pd.read_csv("/content/movie_metadata.csv")
```

```python
movies_df.head()
```

### [Duration] Average lenght

```python
# Average values with two decimal places

print('Average value for "duration" column: %.2f' % movies_df['duration'].mean()) 
```

### [Duration] Missing values and replacing with median

```python
#Looking for duration missing values entries

movies_df[(movies_df['duration'].isna() == True)] 
```

```python
# Median value of duration

movies_df['duration'].median() 
```

```python
# Number of missing values

movies_df['duration'].isna().sum() 
```

```python
# Replacing missing values with median values

movies_df['duration'] = movies_df['duration'].fillna(movies_df['duration'].median()) 
```

```python
# Checking the missing values

movies_df['duration'].isna().sum() 
```

### Movie Duration Category column creation

```python
# Separating the categories of movie duration

movies_df['movie_duration_category'] = ["< 90" if x < 90 else '90-120' if 90 <= x <= 120 else '> 120' for x in movies_df['duration']]
```

```python
movies_df.head()
```

### Year release summary table

```python
# Year summary table

movies_year_stb = movies_df
```

```python
# Separating the null values 

null_index = movies_year_stb[(movies_year_stb['title_year'].isna() == True)]
```

```python
null_index.head()
```

```python
# Dropping the null DataFrame

movies_year_stb.drop(null_index.index, axis=0,inplace=True)
```

```python
# Verifying the nulls

movies_year_stb['title_year'].isna().sum()
```

```python
# Separating the movies released after 2000

movies_2000_stb = movies_year_stb.query('title_year >= 2000')
```

```python
# Changing the column type

movies_2000_stb['title_year'] = movies_2000_stb['title_year'].astype(int)
```

```python
movies_2000_stb.head()
```

```python
# Summary table

duration_year = movies_2000_stb.groupby('title_year')['movie_duration_category'].value_counts()
```

```python
duration_year
```

### Plot category

```python
# Changing the column to string

movies_df_cat = movies_df.astype({'plot_keywords':'string'})
```

```python
movies_df_cat = movies_df_cat.dropna(subset = ['plot_keywords'])
```

```python
# Separating the plot category

movie_plot_category = []

for word in movies_df_cat['plot_keywords']:

  if 'love' in word and 'death' in word:
    movie_plot_category.append('love_and_death')
  elif 'love' in word:
    movie_plot_category.append('love')
  elif 'death' in word:
    movie_plot_category.append('death')
  else:
    movie_plot_category.append('other')
  
```

```python
# Creating tje plot category column

movies_df_cat['movie_plot_category'] = movie_plot_category
```

```python
movies_df_cat['movie_plot_category'].value_counts()
```

```python
movies_df_cat.head()
```

```python
# Verifying the mean score of each category

category_score = movies_df_cat.groupby('movie_plot_category')['imdb_score'].mean()
```

```python
category_score
```

```python
# Score for love category

love_score = category_score.iloc[1]
```

```python
print('Average value for "love" category: %.2f' % love_score)
```

### Budget

```python
# Changing the column to string

movies_df = movies_df.astype({'budget':'string'})
```

```python
# Removing the non numerical string

movies_df['budget'] = movies_df['budget'].str.replace('$', '')
```

```python
# Transforming the column to int

movies_df['budget'] = movies_df['budget'].apply(lambda x: int(float(x)))
```

```python
print('Median value for "budget" column:', movies_df['budget'].median())
```
